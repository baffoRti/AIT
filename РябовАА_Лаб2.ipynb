```python
import pandas as pd
import datetime
```

```python
# Task 1
df = pd.read_csv('data.csv')
df
```

```python
# Task 2
uniq_docs = set()
for cur_str in df['Наименования документов, требованиям которых соответствует средство']:
    if cur_str.startswith('Соответствует требованиям документов: '):
        cur_str = cur_str[len('Соответствует требованиям документов: '):]
        uniq_docs.update(set(cur_str.split(', ')))
print(uniq_docs)
```

```python
# Task 3
uniq_tools = set()
doc_1 = 'ТУ'
doc_2 = 'НМД АРР-2000'
for index, cur_str in df.iterrows():
    if doc_1 in cur_str['Наименования документов, требованиям которых соответствует средство']\
    and doc_2 in cur_str['Наименования документов, требованиям которых соответствует средство']:
        uniq_tools.add(cur_str['Наименование\r\nсредства (шифр)'])
print(uniq_tools)
```

```python
# Task 4
uniq_tools_2 = set()
for index, cur_str in df.iterrows():
    if cur_str['Срок\r\nдействия\r\nсертификата'] != '#Н/Д':
        if datetime.datetime.strptime(cur_str['Срок\r\nдействия\r\nсертификата'], '%d.%m.%Y').date() < datetime.date.today():
            uniq_tools_2.add(cur_str['Наименование\r\nсредства (шифр)'])
print(uniq_tools_2)
```

```python
# Task 5
uniq_tools_3 = set()
for index, cur_str in df.iterrows():
    if cur_str['Срок\r\nдействия\r\nсертификата'] != '#Н/Д'\
    and str(cur_str['Информация об окончании срока технической поддержки, полученная от заявителя']) != 'nan':
        if datetime.datetime.strptime(cur_str['Срок\r\nдействия\r\nсертификата'], '%d.%m.%Y').date() < datetime.date.today():
            if str(cur_str['Информация об окончании срока технической поддержки, полученная от заявителя']) == 'бессрочно'\
            or datetime.datetime.strptime(str(cur_str['Информация об окончании срока технической поддержки, полученная от заявителя']), '%d.%m.%Y').date() >= datetime.date.today():
                uniq_tools_3.add(cur_str['Наименование\r\nсредства (шифр)'])
print(uniq_tools_3)
```

```python
# Task 6
uniq_tools_4 = set()
for index, cur_str in df.iterrows():
    if cur_str['Испытательная\r\nлаборатория'].startswith(('ООО', 'ЗАО', 'ПАО', 'АО')):
        uniq_tools_4.add(cur_str['Наименование\r\nсредства (шифр)'])
print(uniq_tools_4)
```
